## Subscription Types Analysis
We will perform and understand how subscription is reated to other categorical features.

```python
df['Subscription Type'].value_counts().sort_index().plot(kind='bar', position=1)
plt.title('Number of Subscription Type')
plt.xlabel('Subscription Type')
plt.ylabel('Counts')
plt.show()
```

We can see that people are enrolled on basic subscriotion type more comapred to premium and standard plans

```python
value_counts = df[['Country', 'Subscription Type']].value_counts().sort_index().reset_index(name='Count')

# Pivot the data for joined bar graph
pivot_table = value_counts.pivot(index='Country', columns='Subscription Type', values='Count')

# Create the joined bar graph
pivot_table.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Subscription Type Distribution by Country')

# Show the plot
plt.show()
```

```python
value_counts = df[['Subscription Type', 'Gender']].value_counts().sort_index().plot(kind='bar', stacked=True)
plt.title('Gender wise Subscription Type')
```

```python
value_counts = df[['Subscription Type', 'Device']].value_counts().sort_index().plot(kind='bar', position=1)
plt.title('Subscription Type Distribution by Device')
```
